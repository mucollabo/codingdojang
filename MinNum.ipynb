### 1부터 20사이의 어떤 수로도 나누어 떨어지는 가장 작은 수
#### [문제] 1부터 10까지의 어떤 수로도 나누어 떨어지는 가장 작은 수는 2520입니다.
#### 그렇다면 1부터 20까지의 어떤 수로도 나누어 떨어지는 가장 작은 수는 얼마입니까?



```python
n = 1

while True:
  for i in range(1, 21):
    if n%i != 0:
      print(n, i)
      break
  if i == 20:
    break
  n += 1
print(n, i)
    
```
